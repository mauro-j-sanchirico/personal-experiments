```python
import matplotlib.pyplot as plt
import numpy as np


def phi1(x):
    return np.sign(x)


def phi2(x):
    return (1 / 3) * (np.sign(x) + np.sign(x - 1 / 2) + np.sign(x + 1 / 2))


def signal_power(t, signal):
    return np.trapezoid(signal**2, t) * (1 / (2 * np.pi))


t = np.linspace(0, 2 * np.pi, 500)
s = np.sin(t)

e1 = s - phi1(s)
e2 = s - phi2(s)


power = signal_power(t, s)
power_phi1 = signal_power(t, phi1(s))
power_phi2 = signal_power(t, phi2(s))
power_e1_numeric = signal_power(t, e1)
power_e2_numeric = signal_power(t, e2)

power_e1_exact = 3 / 2 - 4 / np.pi
power_e2_exact = 65 / 54 - 4 / (3 * np.pi) - 4 * np.sqrt(3) / (3 * np.pi)

print(f'Power s(t): {power}')
print(f'Power phi1(s): {power_phi1}')
print(f'Power phi2(s): {power_phi2}')
print(f'Power e1(t): {power_e1_numeric}')
print(f'Power e2(t): {power_e2_numeric}')
print(f'Exact Power e1(t): {power_e1_exact}')
print(f'Exact Power e2(t): {power_e2_exact}')
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 8), sharex=False)

axes[0].plot(t, np.sin(t), color='C0')
axes[0].set_title('sin(t)')
axes[0].grid(True)

x_sign = np.linspace(-1.5, 1.5, 500)

axes[1].plot(x_sign, phi1(x_sign), color='C1')
axes[1].set_title('phi1(x)')
axes[1].grid(True)

axes[2].plot(x_sign, phi2(x_sign), color='C2')
axes[2].set_title('phi2(x)')
axes[2].grid(True)

for axis in axes:
    axis.set_ylabel('value')
axes[2].set_xlabel('x')

fig.tight_layout()
plt.show()
```
